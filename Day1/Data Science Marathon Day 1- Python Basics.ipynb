### Lists, tuples, sets, and dictionaries

These are python data structures that stores a sequence of values

#### i. Lists

-Contained in square brackets<br>
-Contain several values of different data types<br>
-Can even contain a list (list of a list)

```python
list=[1,2,3,4,5] #list of integers
list
```

Basic list methods:<br>
                    • Concatenation<br>
                    • Repetition<br>
                    • Length<br>
                    • Appending<br>
                    • Sorting<br>
                    • Indexing

```python
#list Concatenation- use plus sign
list1=[1,2,3,4,5]
list2=[1,2,0]

#Concatenate the 2 lists
list1+list2
```

```python
#List repetition- use multiplication sign
list1=[1,2,0]

#Repeat the list twice
list1*2
```

```python
#Length of a lsit- use len() function. Returns no of elements in the list
list1=[1,2,0]

#Get lenth of the list
len(list1)
```

```python
#List appending- use append() function. Adds an element at the end of a list
list1=[1,2,0]

#Append 2 to the list
list1.append(2)
list1
```

```python
#Sorting a list- Done with sort and sorted functions- arranges the elements in ascending manner
list1=[3,2,0]
list1.sort()
list1
```

```python
#To sort in descending manner, add reverse keyword
list1=[2,3,0]
list1.sort(reverse=True)
list1
```

```python
#Indexing
list1=[3,2,0]
list1[0] #list element at index 0
```

```python
list1[0:-2]
```

```python
list1[len(list1)-1]#gets last element in the list
```

```python
tuple1 = (2, 3)
tuple1
```

```python
#Convert a tuple to a list
tuple1 = (2, 3)
list(tuple1)
```
