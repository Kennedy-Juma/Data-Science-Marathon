### Loops and comprehensions

Loops- Used for iteration. Are important because they enable us to step through lists/ other data structures methodically, one element at a time.

Examples of loops- for and while loops

#### For Loop

looping through a list using for loop

```python
a_list = [1, 2, 3]
for element in a_list:
    print(element) #Prints elements of the list one at a time
```

Continue and break statements when using for loops

```python
#The break keyword will end a loop
for element in a_list:
    print(element)
    break
```

```python
#continue keyword will immediately move on to the next iteration
for element in a_list:
    print(element)
    continue
```

Commonly used functions with loops- range() and len()

```python
for i in range(len(a_list)):
    print(i)
```

```python
for i in range(1,7,1):
    print(i, end=' ')
```

```python
a_list = [1, 2, 3]
for index, element in enumerate(a_list): #Using inbuilt function enumerate. Returns a tuple
    print(index, element)
```

```python
empty_list=[]
for element in range(5):
    empty_list.append(element)
```

```python
empty_list
```

```python
#Dictionary comprehension using .items() keyword
a_dict = {'books': 1, 'magazines': 2, 'articles': 7}
for key, value in a_dict.items():
    print(f'{key}:{value}')
```

### Booleans and conditionals

Booleans are data types that take the binary values of True (1) or False (0).

```python
my_number = 11
my_number > 10
```

Boolean values are often used in if... else statements

if/else statement in Python gives us options for what to do with our code if certain conditions are met.

```python
#if else with booleans
books_read = 12
if books_read < 10:
    print("You have only read a few books.")
elif books_read >= 12:
    print("You've read lots of books!")
else:
    print("You've read 10 or 11 books.")
```
